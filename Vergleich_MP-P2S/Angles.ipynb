```python
import pandas as pd
```

```python
def print_angle_graphs(file_path):
    with open(file_path, "r") as f:
        for i, line in enumerate(f):
            if line.strip() == "endheader":
                header_line = i + 1
                break

    df = pd.read_csv(
        file_path,
        sep=r'\s+',
        skiprows=header_line
    )

    p = df.plot(y=[i for i in range(1, len(df.head()) - 1)], xlabel="Frame", ylabel="Angle in Degrees")
    p.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    p.plot()
```

```python
print_angle_graphs("MediaPipe/IKResult_mp.mot")
```

```python
print_angle_graphs("Pose2Sim/IKResult_p2s.mot")
```
